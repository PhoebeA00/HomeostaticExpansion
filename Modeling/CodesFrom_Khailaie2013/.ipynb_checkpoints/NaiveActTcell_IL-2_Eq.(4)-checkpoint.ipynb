```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Conclusion
# INITIATION OF AN IMMUNE RESPONSE REQUIRES A MINIMUM HOMEOSTATIC POPULATION OF NA√èVE T CELLS AND ANTIGEN STIMULATION
```

```python
#function that returns dy/dt
def model(z,t,co):
    # Initial conditions
    T = z[0]
    I = z[1]
    N = z[2]
    # Constant Values
    a = co[0]
    b = co[1]
    d = co[2]
    e = co[3]
    f = co[4]
    g = co[5]
    beta = co[6]
    
    dNdt = 0*N - g*N - beta * N
    dTdt = a*T * I + beta*T - b*t
    dIdt = d*T - e*I*T - f*I
    return [dTdt, dIdt, dNdt]
```

```python
# initial condition
# z0 = [T,I,N_0]
z0 = [1,0.5,200]

# Constant values co = [a,b,d,e,f,g,beta]
co = [0.4,0.1,0.01,0.01,1,0.1,0.5]
# time points
t = np.linspace(0,10)
```

```python
# solve ODE
z = odeint(model,z0,t, args=(co,))

T = z[:,0]
I = z[:,1]
N = z[:,2]

fig = plt.figure(figsize=(20,10))

ax1 = fig.add_subplot(131)
ax1.plot(t,N, 'b-')
ax1.set_xlabel('time')
ax1.set_ylabel('Naive T')
ax1.set_title('Naive T')

ax2 = fig.add_subplot(132)
ax2.plot(t,T,'b-')
ax2.set_xlabel('time')
ax2.set_ylabel('Activated T Cells')
ax2.set_title('Activated T Cells')


ax3 = fig.add_subplot(133)
ax3.plot(t,I,'b-')
ax3.set_xlabel('time')
ax3.set_ylabel('IL-2 Cytokines')
ax3.set_title('IL-2 Cytokines')
```
