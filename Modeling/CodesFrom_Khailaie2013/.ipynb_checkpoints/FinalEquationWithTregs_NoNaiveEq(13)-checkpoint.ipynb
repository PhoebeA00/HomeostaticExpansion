```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
# Questions

# what does a represent, and why is it ok to have a on both T and R
    # Because the expression of Tregs is relative to naive T cells

# Notes
# the control the renewal rate of tregs by making relative to the renewal rate of Naive T cells
# The relative proliferation rate of activated Tregs and activated T cells is controlled by the parameter epsilon 
    # - IL-2 dependent proliferation is relative to activated T cells
```

```python
#function that returns dy/dt
def model(z,t,co):
    # Initial conditions
    T = z[0]
    I = z[1]
    R = z[2]
    N = z[3]
    
    # Constant Values
    # co = [a,b,d,e,f,g,beta,c, gamma, epsilon,lambd]
    a = co[0]
    b = co[1]
    d = co[2]
    e = co[3]
    f = co[4]
    g = co[5]
    beta = co[6]
    c = co[7]
    gamma = co[8]
    epsilon = co[9]
    N_0 = 4
    lambd = co[10]#0.006, 0.02
    N_hat = lambd*N_0 
    
    
    dTdt = a*I * T - b*t - c*T**2 - gamma*R*T + beta*T
    dRdt = epsilon*a*I*R - b*R + beta*N_hat
    dIdt = d*T - e*I*(T + R) - f*I
    dNdt = N_0 - g*N - beta * N
    
    return [dTdt, dIdt, dRdt, dNdt]
```

```python
# initial condition
# z0 = [T,I,N_0]
z0 = [5,0.5,1,200]

# Constant values co = [a,b,d,e,f,g,beta,c]
a = 0.4
b = 0.1
d = 0.01
e = 0.01
f = 1
g = 0.01 # No value given
beta = 0.5 # [0, inf]
c = 10**-5
gamma = 0.1 
epsilon = 0.6
lambd = 0.006 #0.006, 0.02
co = [a,b,d,e,f,g,beta,c, gamma, epsilon,lambd]
# co = [0.4,0.1,0.01,0.01,1,0.01,0.5, 10**-5]
```

```python
# time points
t = np.linspace(0,100)
# solve ODE
z = odeint(model,z0,t, args=(co,))

T = z[:,0]
I = z[:,1]
R = z[:,2]
N = z[:,3]

fig = plt.figure(figsize=(20,10))

ax4 = fig.add_subplot(221)
ax4.plot(t,N, 'b-')
ax4.set_xlabel('time')
ax4.set_ylabel('Naive')
ax4.set_title('Naive')

ax2 = fig.add_subplot(222)
ax2.plot(t,T,'b-')
ax2.set_xlabel('time')
ax2.set_ylabel('Activated T Cells')
ax2.set_title('Activated T Cells')

ax1 = fig.add_subplot(223)
ax1.plot(t,R, 'b-')
ax1.set_xlabel('time')
ax1.set_ylabel('Tregs')
ax1.set_title('Tregs')

ax3 = fig.add_subplot(224)
ax3.plot(t,I,'b-')
ax3.set_xlabel('time')
ax3.set_ylabel('IL-2 Cytokines')
ax3.set_title('IL-2 Cytokines')
```
