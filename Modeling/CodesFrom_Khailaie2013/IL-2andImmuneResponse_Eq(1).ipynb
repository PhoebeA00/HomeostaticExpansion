```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
```

```python
#function that returns dy/dt
def model(z,t,co):
    # Initial conditions
    T = z[0]
    I = z[1]
    # Constant Values
    a = co[0]
    b = co[1]
    d = co[2]
    e = co[3]
    f = co[4]
    
    dTdt = a*T * I + b*T
    dIdt = d*T - e*I*T - f*I
    return [dTdt, dIdt]
```

```python
# initial condition
z0 = [66,0.5]

# Constant values co = [a,b,d,e,f]
co = [0.4,0.1,0.01,0.01,1]
# time points
t = np.linspace(0,10)
```

```python
# solve ODE
z = odeint(model,z0,t, args=(co,))

td = z[:,0]
il = z[:,1]
```

```python
f, axarr = plt.subplots(1,3, figsize=(20,4))
axarr[0].plot(t,td,'b-')
axarr[0].set_xlabel('time')
axarr[0].set_ylabel('CD4Tcells')
axarr[0].set_title('CD4 T cells')
axarr[1].plot(t,il,'b-')
axarr[1].set_xlabel('time')
axarr[1].set_ylabel('CD4 T cells')
axarr[1].set_title('IL-2 Cytokines')
axarr[2].plot(il,td,'b-')
axarr[2].set_xlabel('IL-2 Cytokines')
axarr[2].set_ylabel('CD4 T cells')
axarr[2].set_title('CD4 T cells vs. IL-2 Cytokines')
```
