```python
import pandas as pd
import seaborn as sns
from collections import Counter
```

```python
pop = pd.read_csv('~/my.work/PhD/Homestatic Expansion Project/ModelData/AfterCalculations.csv')
act = pd.read_csv('/home/jon/my.work/PhD/Homestatic Expansion Project/ModelData/ActivatedCD4pop.csv')
#Making the naive data
act['NaiveT'] = act['CD4CT'] - act['ActivatedCD4CT']
```

```python
#Getting the spleen data
Spop = pop.loc[pop['Organ'] == 'Spleen']
Sact = act.loc[act['Organ'] == 'Spleen']
```

```python
#Split in Ages 0-9
Spop09 = Spop.loc[Spop['intage'] < 12]
Sact09 = Sact.loc[Sact['Age'] < 12]

#Split in Ages 12-A
Spop12A = Spop.loc[Spop['intage'] > 9]
Sact0912A = Sact.loc[Sact['Age'] > 9]
```

```python
#Plotting
lm1 = sns.lmplot(y="CD4CT", x="intage", hue="Genotype", data=Spop09)
lm1f = lm1.fig
lm1f.suptitle("CD4 count", fontsize=12)

lm2 = sns.lmplot(y="ActivatedCD4CT", x="Age", hue="Genotype", data=Sact09)
lm2f = lm2.fig
lm2f.suptitle('Activated CD4 T cells', fontsize = 12)

lm3 = sns.lmplot(y="NaiveT", x="Age", hue="Genotype", data=Sact09)
lm3f = lm3.fig
lm3f.suptitle('Naive T Cells')
```
